```python
import pandas as pd
from numpy.random import randint
import matplotlib.pyplot as plt
InputFileName = "/Users/mxm/Google Drive/Masterstudium/Inhalte/Master Thesis/GitHubRepo/agile-in-government/Analysis/2_Data_Preprocessing/DATA/CSVs/UK/General_v1.csv"

df=pd.read_csv(inputFileName, parse_dates=["final_date"])
grouped = df[df.final_date < '2019-12-31 23:59:59']["final_date"].groupby([df["final_date"].dt.year]).count()
#grouped = df.groupby([final_date)
rowlength = grouped.ngroups/2                         # fix up if odd number of groups
fig, axs = plt.subplots(figsize=(9,4), 
                        nrows=2, ncols=rowlength,     # fix as above
                        gridspec_kw=dict(hspace=0.4)) # Much control of gridspec

targets = zip(grouped.groups.keys(), axs.flatten())
for i, (key, ax) in enumerate(targets):
    ax.plot(grouped.get_group(key))
    ax.set_title('a=%d'%key)
ax.legend()
plt.show()
```
